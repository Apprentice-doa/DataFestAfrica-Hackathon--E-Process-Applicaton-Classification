```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

print("Sucessful")
```

```python
trainset = pd.read_csv("Data/Datadoa.csv")
testset = pd.read_csv("Data/test.csv")
```

```python
trainset.head()
```

```python
trainset.drop("Entry_id", axis = 1, inplace = True)
X = trainset.drop("e_signed", axis = 1)
y = trainset["e_signed"]
```

```python
n_estimators = [int(x) for x in np.linspace(start = 5, stop=20, num = 1)]
max_features = ['sqrt','auto']
max_depth =[2,4]
min_samples_split =[2,4]
bootstrap = [True,False]
```

```python
param_grid = {'n_estimators':n_estimators,
              'max_features': max_features,
              'max_depth': max_depth,
              'min_samples_split':min_samples_split,
              'bootstrap':bootstrap,
    
}
from sklearn.model_selection import GridSearchCV
```

```python
from sklearn.ensemble import RandomForestClassifier 

rfmodel = RandomForestClassifier()

rfmodel_Grid = GridSearchCV(estimator = rfmodel, param_grid = param_grid, cv =3, verbose = 2, n_jobs = 4)
```

```python
rfmodel_Grid.fit(X,y)
```

```python
rfmodel_Grid.best_estimator_
```

```python
from sklearn.model_selection import cross_val_score
#for feature in X.columns:
scores = cross_val_score(rfmodel_Grid.best_estimator_,X,y,scoring = "roc_auc", cv = 10)
scores.mean()
```

```python
from sklearn.preprocessing import StandardScaler
scale = StandardScaler()

X_scaled = scale.fit_transform(X)
```

```python
import catboost as ctb


ctbmodel = ctb.CatBoostClassifier()

ctbmodel.fit(X_scaled,y, silent = True)

```

```python
from sklearn.model_selection import cross_val_score
scores = cross_val_score(ctbmodel,X_scaled,y,scoring = "roc_auc", cv = 5)
scores.mean()
```
